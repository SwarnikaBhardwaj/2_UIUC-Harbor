```python
def load_model(model_name):
    print(f"\n--- Loading {model_name} ---")
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(
        model_name,
        # torch_dtype=torch.float16,
        dtype=torch.float16,
        device_map="auto",
        trust_remote_code=True
    )

    return tokenizer, model
```

```python
test_prompt = "Extract structured JSON from: 'Math tutoring for $20/hour."
```

```python
def generate_response(tokenizer, model, prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    start = time.time()

    outputs = model.generate(
        **inputs,
        max_new_tokens=150
    )

    end = time.time()

    response = tokenizer.decode(outputs[0][inputs['input_ids'].shape[-1]:], skip_special_tokens=True)

    print("Response:\n", response)
    print("Time:", round(end - start, 2), "seconds")
```

```python
for category, model_list in models.items():
    for model_name in model_list:
        try:
            tokenizer, model = load_model(model_name)
            generate_response(tokenizer, model, test_prompt)
            # Clearing memory for "Matrix Testing"
            del model
            del tokenizer
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
            gc.collect()

        except Exception as e:
            print(f"Failed to load {model_name}")
            print(e)
```

```python
test_prompts = {
    "Zero-shot": "Classify the following request into a service category:\n'I need someone to design a logo for my startup.'",

    "One-shot": """Example:
'I need a math tutor.' → Education

Now classify:
'I need someone to design a logo for my startup.'""",

    "Few-shot": """Examples:
'I need a math tutor.' → Education
'Looking for a photographer.' → Creative
'Need help moving furniture.' → Physical Labor

Now classify:
'I need someone to design a logo for my startup.'"""
}
```

```python
import torch
import time
import gc
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig

quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)
def test_model(model_id, prompts):
    print(f"\nLoading {model_id}")

    try:
        start_time = time.time()

        tokenizer = AutoTokenizer.from_pretrained(model_id)
        # model = AutoModelForCausalLM.from_pretrained(model_id)
        # model.to(device)
        model = AutoModelForCausalLM.from_pretrained(
            model_id,
            quantization_config=quant_config, # This is the magic line
            device_map="auto",                # This handles GPU/CPU offloading
            low_cpu_mem_usage=True,
            trust_remote_code=True
        )

        load_time = time.time() - start_time
        print(f"Load time: {load_time:.2f} seconds")

        for prompt_type, prompt in prompts.items():
            print(f"\n--- {prompt_type} ---")

            inputs = tokenizer(prompt, return_tensors="pt").to(device)
            outputs = model.generate(
                **inputs,
                max_new_tokens=50,
                do_sample=False
            )

            response = tokenizer.decode(outputs[0], skip_special_tokens=True)
            print(response)
            print("-" * 50)

    except Exception as e:
        print(f"Failed to load {model_id}: {e}")

    finally:
        # Clean memory
        try:
            del model
            del tokenizer
        except:
            pass

        if torch.backends.mps.is_available():
            torch.mps.empty_cache()
        gc.collect()
```

```python
for category, models in model_matrix.items():
    print(f"\n========================")
    print(f" CATEGORY: {category}")
    print(f"========================")

    for model_id in models:
        test_model(model_id, test_prompts)
```

Part 4: Evaluation, Using Route A (Prompt Engineering)


Evaluation of Prompting Strategies
Our team tested Qwen2.5-0.5B-Instruct and found that a corelation between the prompt complexity and output reliability. Using a zer-shot condition, the model did not adhere to the requested schema, and also resulted in repetition loops or repeating the input texts. This resulted in a low reliability score of 2/10

A one-shot example improved the structural attempt, but the model still resulted with filler words that broke the json parser. This earned a score of 5/10

This most successful method was few-shot prompting which had consistent results, and valid jscon structure. It also had a reliability score of 8/10 and generated an output in about 24 seconds. Most many-shot prompts had lsightly higher accuracy but was ultimately rejected because it required high numbers of token and made it less ideal for implementing into our project.

Top Performing Strategy: Few-shot Prompting.

Selected Model: Qwen2.5-0.5B (for Ultra-Light local testing) and Mistral-7B (quantized) for production-level reasoning.

Conclusion: By using few-shot prompts, we can achieve high-quality data extraction for our an audit without needing full model fine-tuning.
